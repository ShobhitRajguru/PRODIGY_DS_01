```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your dataset from a CSV file
df = pd.read_csv('Population_data.csv')
df
```

```python
# Plotting
plt.figure(figsize=(14, 8))
# Bar chart
plt.bar(df['Location'], df['Male'], label='Male', color='skyblue', alpha=0.7)
plt.bar(df['Location'], df['Female'], bottom=df['Male'], label='Female', color='lightcoral', alpha=0.7)
# Add labels and title
plt.xlabel('Location')
plt.ylabel('Population Distribution')
plt.title('Population Distribution by Sex')
plt.legend()
# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
```
